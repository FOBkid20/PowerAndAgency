### Find proper nouns

```python
def find_propers(review):
    nlp = spacy.load("en_core_web_sm")
    doc = nlp(review)
    propns = []
    for token in doc:
    #     print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_,
    #             token.shape_, token.is_alpha, token.is_stop)
        if (token.pos_ == "PROPN"):
            propns.append(token.text)
    return(propns)
```

# Dependency

```python
def avt_parse(review):
    doc = nlp(review)
    # doc = nlp("I was a hippie age 22 in 1965, have seen the play 5 times, have 2 versions of the music, and have read the script many times.")

    objects = ["dobj", "amod", "nummod", "acomp"]
    all_avts = []
    for possible_subject in doc:
        avt = {}
        if possible_subject.dep_ == "nsubj" and (possible_subject.head.pos_ == "VERB"):
    #     if possible_subject.dep_ == "nsubj" and (possible_subject.head.pos_ == "VERB" or possible_subject.head.pos_ == "AUX"):
        #passives = nsubjpass    
            agent = possible_subject
            verb = possible_subject.head
            possible_themes = verb.children
            found = False
            for possible_theme in possible_themes:
                if possible_theme.dep_ in objects:
                    theme = possible_theme
                    found = True
                    # TODO: probably should account for multiple themes not just one
            avt["Agent"] = agent.text
            avt["Verb"] = verb.lemma_
            if(found): 
                avt["Theme"] = theme.text
            all_avts.append(avt)

#     print(all_avts)
    return(all_avts)
```

### Cull by gender

```python
def get_gendered_avts(all_avts, name_gender_dict):
    gendered_words = name_gender_dict.keys()
    gendered_avts = []
    for avt in all_avts:
        if avt["Agent"] in gendered_words:
            avt["Agent"] = name_gender_dict[avt["Agent"]]
            gendered_avts.append(avt)
        if "Theme" in avt.keys():
            if avt["Theme"] in gendered_words:
                avt["Theme"] = name_gender_dict[avt["Theme"]]
                gendered_avts.append(avt)

#     print(gendered_avts)
    return(gendered_avts)
```

```python
def add_to_data(data, key, val):
    if key in data.keys():
        lst = data[key]
        lst.append(val)
        data[key] = lst
    else:
        data[key] = [val]
    return data
    
    
```
